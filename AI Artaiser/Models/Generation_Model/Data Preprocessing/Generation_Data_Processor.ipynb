# **3. Download images from URL**
Download the dataset from the web into google drive

```python
for i in image_df.index:
    image_url = image_df['URL'][i]
    filename = '/content/drive/MyDrive/AI_Artathon/Phase_2/Models/Prediction_Model/Dataset/SuperRare_Original/{}.png'.format(i)
    print("Image [{}]:".format(i))
    try:
        r = requests.get(image_url, stream = True)
    except:
        pass
    if r.status_code == 200:
      r.raw.decode_content = True
      with open(filename,'wb') as f:
        shutil.copyfileobj(r.raw, f)
        print('Image sucessfully Downloaded')        
    else:
      print('Image Couldn\'t be retreived')
```
